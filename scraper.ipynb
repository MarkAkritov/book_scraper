```python
import pandas as pd
from itertools import chain
import time
```

```python
urls = ["https://www.amazon.com/gp/bestsellers/{0}/books/ref=zg_bsar_pg_2?ie=UTF8&pg={1}".format(y,i) for y in range(1995, 2019) for i in range(1,3)]
```

```python
from selenium import webdriver
```

```python
years = {}
for url in urls:
    hrefs = []
    titles = []
    authors = []
    stars = []
    reviews = []
    cover_type = []
    prices = []
    publishers = []
    genres = []
    browser = webdriver.Chrome()
    browser.get(url)
    hrefs.extend([i.get_attribute('href') for i in browser.find_elements_by_xpath('//span[@class="aok-inline-block zg-item"]/a[1]')])
    for href in hrefs:
        browser1 = webdriver.Chrome()
        browser1.get(str(href))
        try:
            publishers.extend([i.text.replace('Publisher: ', '') for i in browser1.find_elements_by_xpath("//div[@class='content']/ul/li") if 'Publisher' in i.text])
            genres.extend([i.text.split('\n')[1].split('>')[1] for i in browser1.find_elements_by_xpath("//div[@class='content']/ul/li") if "Amazon Best Sellers Rank" in i.text])
        except IndexError:
            publishers.extend(' ')
            genres.extend(' ')
        browser1.close()
        time.sleep(2)
    titles.extend([i.get_attribute('alt') for i in browser.find_elements_by_xpath('//div[@class="a-section a-spacing-small"]/img')])
    authors.extend([i.text for i in browser.find_elements_by_xpath("//div[@class='a-row a-size-small']/*[self::span[@class='a-size-small a-color-base']|self::a]")])
    stars.extend([i.get_attribute('title')[:3] if i is not None else " " for i in browser.find_elements_by_xpath("//div[@class='a-icon-row a-spacing-none']/a[@class='a-link-normal']")])
    reviews.extend([i.text if i is not None else " " for i in browser.find_elements_by_xpath("//div[@class='a-icon-row a-spacing-none']/a[@class='a-size-small a-link-normal']")])
    cover_type.extend([i.text for i in browser.find_elements_by_xpath("//div[@class='a-row a-size-small']/span[@class='a-size-small a-color-secondary']")])
    prices.extend([i.text for i in browser.find_elements_by_xpath("//span[@class='a-size-base a-color-price']/span[@class='p13n-sc-price']")])
    years[url.split('/')[5]+'_'+url[-1]] = {
        'titles': titles,
        'authors': authors,
        'stars': stars,
        'reviews': reviews,
        'cover': cover_type,
        'prices': prices,
        'publishers': publishers,
        'genres': genres
    }    
    browser.close()
    
```
